```python
def find_recipes_by_ingredients(ingredients):
    matching_recipes = RAW_recipes[RAW_recipes['ingredients'].apply(lambda x: all(ingredient in x for ingredient in ingredients))]
    
    return matching_recipes[['name', 'id']]

find_recipes_by_ingredients(['pasta', 'salad', 'mushroom'])
```

```python
def count_and_rating(recipes):
    # Append 'count' column
    recipes['count'] = recipes['id'].apply(lambda x: RAW_interactions[RAW_interactions['recipe_id'] == x].shape[0])
    
    # Append 'rating' column
    recipes['rating'] = recipes['id'].apply(lambda x: RAW_interactions[RAW_interactions['recipe_id'] == x]['rating'].mean())
    
    # Sort by 'count' column in descending order, and then by 'rating' column in descending order
    recipes_sorted = recipes[['name', 'count', 'rating']].sort_values(by=['count', 'rating'], ascending=[False, False])
    recipes_sorted.columns = ['Recipe Name', 'Reviews (#)', 'Average Rating']
    
    return recipes_sorted

    
count_and_rating(find_recipes_by_ingredients(['pasta', 'salad', 'mushroom']))
    
```
